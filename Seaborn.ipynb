### Correlation with Heatmap

Correlation heatmap uses colored cells on monorchromatic scale between two discrete dimensions.

```python
df.corr()
#Here only integer, float type variables are shown 
```

```python
sns.heatmap(df.corr())
```

```python
#Join plot- It allow us to study the relationship between two integer features.
```

```python
#Bivariate analysis
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
#Here above jointplot shows that majority of tip given is upto 2 and total bill between 10 to 20
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

```python
#Paiir plot is also used as scatterplot where one vaiable is matched with another variable
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
sns.pairplot(df,hue='smoker')
```

# Distplot-

It checks the distribution of the column feature


```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

## Categorical Plots

*  boxplot
*  violinplot
*  countplot
*  bar plot

```python
sns.countplot('sex',data=df)
```

```python
sns.countplot(y='smoker',data=df)
```

```python
#barplot- We can specifiy the x axis and y axis column

sns.barplot(x='sex',y='total_bill',data=df)
```

```python
sns.barplot(x='smoker',y='total_bill',data=df)
```

```python
#Box plot

sns.boxplot('smoker','total_bill',data=df)
```

```python
sns.boxplot('day','total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df)
```

```python
# categorize my data based on some other categories

sns.boxplot(x="total_bill", y="day", hue="smoker",data=df)
```

```python
#Violinplot is to see the distibution of data in the form of  kernel density estimation
```

```python
sns.violinplot(x="total_bill", y="day", data=df,palette='rainbow')
```
